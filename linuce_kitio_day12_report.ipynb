### Exercise 1:
Create a simple line plot for `y = x` for `x` values ranging from 0 to 10.

```python
# Exercise 1
```

---

````python

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # for 3D plotting
````

````python
x = np.linspace(0, 10)
y = x

# create a figure/canvas of desired size
plt.figure(figsize=(8, 6))

# plot values; with a color `red`
plt.plot(x, y, 'r')

# give labels to the axes
plt.xlabel('x')
plt.ylabel('y')

# give a title to the plot
plt.title(r"Plot of $y=x$")
plt.show()
````

### Exercise 4:
Plot a sine wave using Matplotlib for `x` values between 0 and 2.

```python
# Exercise 4
```

---

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)


plt.figure(figsize=(10, 4))
plt.plot(x, y, label='Sine Wave')
plt.title('Sine Wave for x values between 0 and 2')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.show()
````

### Exercise 7:
Plot two line plots on the same graph: `y1 = x^2` and `y2 = x^3` for `x` ranging from 0 to 10.

```python
# Exercise 7
```

---

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)

# Compute y1 and y2 values
y1 = x ** 2
y2 = x ** 3

# Create the plot
plt.figure(figsize=(8, 4))
plt.plot(x, y1, label='$y = x^2$', color='blue')
plt.plot(x, y2, label='$y = x^3$', color='green')
plt.show()
````

### Exercise 9:
Create a subplot with 2 rows and 1 column. Plot `y1 = x^2` in the first subplot and `y2 = x^3` in the second.

```python
# Exercise 9
```

---

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)

y1 = x ** 2
y2 = x ** 3

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8))


ax1.plot(x, y1, color='blue')
ax1.set_title('$y = x^2$')
ax1.set_xlabel('x')
ax1.set_ylabel('y')


ax2.plot(x, y2, color='green')
ax2.set_title('$y = x^3$')
ax2.set_xlabel('x')
ax2.set_ylabel('y')


plt.tight_layout()
plt.show()
````

### Exercise 10:
Use a different color and linestyle for each line in a plot of three functions: `y1 = x`, `y2 = x^2`, and `y3 = x^3`.

```python
# Exercise 10
```

---

````python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 10, 100)


y1 = x
y2 = x ** 2
y3 = x ** 3


plt.figure(figsize=(8, 6))


plt.plot(x, y1, label='$y = x$', color='blue', linestyle='-')


plt.plot(x, y2, label='$y = x^2$', color='red', linestyle='--')


plt.plot(x, y3, label='$y = x^3$', color='green', linestyle=':')


plt.title('Line Plots for $y = x$, $y = x^2$, and $y = x^3$')
plt.xlabel('x')

plt.ylabel('y')


plt.legend()


plt.show()
````

### Exercise 17:
Create a plot with multiple subplots in a 2x2 grid. Plot `y1 = x`, `y2 = x^2`, `y3 = x^3`, and `y4 = x^4`.

```python
# Exercise 17
```

---

````python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 10, 100)


y1 = x
y2 = x ** 2
y3 = x ** 3
y4 = x ** 4


fig, axs = plt.subplots(2, 2, figsize=(10, 10))


axs[0, 0].plot(x, y1, color='blue')
axs[0, 0].set_title('$y = x$')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')

axs[0, 1].plot(x, y2, color='red')
axs[0, 1].set_title('$y = x^2$')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')

axs[1, 0].plot(x, y3, color='green')
axs[1, 0].set_title('$y = x^3$')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')

axs[1, 1].plot(x, y4, color='purple')
axs[1, 1].set_title('$y = x^4$')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')


plt.tight_layout()
plt.show()
````
